```python
from datetime import date, datetime, timedelta
with open("ossec.log.txt") as f:
    text=f.readlines()
#print(text)
#print(len(text))
pattern = "(?P<year>\d{4})-(?P<month>\d{2})-(?P<day>\d{2})"
for line in text:
    m = re.search(pattern,line)
    # print(result.group())
    print (m.group("year"),m.group("month"),m.group("date"))
```
